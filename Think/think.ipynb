分类和回归

分类:
给定一组数据,训练模型,最后给定不知分类的数据,希望通过这个模型给出预测的分类

回归
给定一组数据,训练模型,最后给定不知结果的数据,希望通过这个模型可以预测结果

区别:
分类的结果是离散的
回归的结果是连续的
