而在假定的过程中 y = wx + w0 也就是多元函数,其中有截距和斜率的概念,因为是平方,从而是二次曲线,要求曲线的最小值,那就是导数为0的地方,但是计算机应该如何求导呢?

按照已知的内容,直接使用的是梯度下降方法, 梯度就是某一点的最大变化率,而我们主要就是让这个函数取得最小值,从而是可以使用这个方法的
